```python
import os
import pandas as pd
from PIL import Image
import numpy as np

BASE_DIR = "/home/shared-data/AnnotierteBilder"
CSV_PATH = "/home/javid/corrosion-detector/fully_valid_samples.csv"
OUT_IMG_DIR = "../data/images"
OUT_MASK_DIR = "../data/masks"

os.makedirs(OUT_IMG_DIR, exist_ok=True)
os.makedirs(OUT_MASK_DIR, exist_ok=True)

df = pd.read_csv(CSV_PATH)

print(f"Total valid samples: {len(df)}")

for idx, row in df.iterrows():
    rgb_path = row["rgb"]
    dmg_path = row["dmg"]

    sample_name = row["sample"]

    # Load RGB
    img = Image.open(rgb_path).convert("RGB")
    img.save(f"{OUT_IMG_DIR}/{sample_name}.jpg")

    # Load damage mask correctly (fix RGBA issue)
    dmg_mask = Image.open(dmg_path).convert("L")       # <-- FIXED (IMPORTANT)
    dmg_mask = np.array(dmg_mask)

    # Threshold (FIXED)
    binary_mask = np.where(dmg_mask > 30, 1, 0).astype(np.uint8)

    # Save final mask
    Image.fromarray(binary_mask * 255).save(f"{OUT_MASK_DIR}/{sample_name}.png")

print("DONE: All images and masks prepared successfully!")
```
