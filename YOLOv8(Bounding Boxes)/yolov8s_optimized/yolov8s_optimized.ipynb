```python
from ultralytics import YOLO
import os
```

<div class="alert alert-block alert-info">

----------
---------
# <b> 2. Defining Paths</b> 

--------------
----------------
</div>

```python
DATA_YAML = "/home/javid/corrosion_detector_part2/dataset/data.yaml"
RUNS_DIR = "/home/javid/corrosion_detector_part2/runs"
MODEL_NAME = "yolov8s_optimized"
MODEL_PATH = os.path.join(RUNS_DIR, MODEL_NAME, "weights", "best.pt")
```

<div class="alert alert-block alert-info">

----------
---------
# <b> 3. Loading the trained YOLOv8 model</b> 

--------------
----------------
</div>

```python
best_model = YOLO(MODEL_PATH)
```

<div class="alert alert-block alert-info">

----------
---------
# <b> 4. Evaluating the model on the test dataset</b> 

--------------
----------------
</div>

```python
metrics = best_model.val(data=DATA_YAML)
print("\n Validation Metrics:")
for k, v in metrics.results_dict.items():
    print(f"{k:20}: {v:.4f}")
```

<div class="alert alert-block alert-info">

----------
---------
# <b> 6. Running inference on test images and saving predictions</b> 

--------------
----------------
</div>

```python

best_model.predict(
    source="/home/javid/corrosion_detector_part2/dataset/images/test",
    project=RUNS_DIR,
    name="yolov8s_optimized_predictions",
    save=True,
    imgsz=800,
    conf=0.25
)
```
